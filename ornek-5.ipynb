```python
import numpy as np
import matplotlib.pyplot as plt

# Hedef (gerçek) parametreler
a_true = 2
b_true = 1
c_true = np.pi / 4
d_true = 3

# Giriş verisi
x = np.linspace(0, 2 * np.pi, 100)

# Hedef (gerçek) çıktı
y_true = a_true * np.sin(b_true * x + c_true) + d_true

# Tahmin edilen parametrelerin başlangıç değerleri
a = 1
b = 2
c = 0
d = 1

# Öğrenme hızı
alpha = 0.01

# Eğitim döngüleri sayısı (epoklar)
epochs = 1000

# Eğitim döngüsü
for epoch in range(epochs):
    # Tahmin edilen çıktı
    y_pred = a * np.sin(b * x + c) + d

    # Kayıp fonksiyonu
    loss = np.mean((y_true - y_pred) ** 2)

    # Kayıp fonksiyonunun türevleri
    da = np.mean(2 * (y_pred - y_true) * np.sin(b * x + c))
    db = np.mean(2 * (y_pred - y_true) * a * x * np.cos(b * x + c))
    dc = np.mean(2 * (y_pred - y_true) * a * np.cos(b * x + c))
    dd = np.mean(2 * (y_pred - y_true))

    # Parametreleri güncelle
    a -= alpha * da
    b -= alpha * db
    c -= alpha * dc
    d -= alpha * dd

# Eğitim sonrası tahmin edilen çıktı
y_pred = a * np.sin(b * x + c) + d

# Gerçek ve tahmin edilen çıktıları çizdir
plt.plot(x, y_true, label='Gerçek')
plt.plot(x, y_pred, label='Tahmin')
plt.legend()
plt.show()
```
